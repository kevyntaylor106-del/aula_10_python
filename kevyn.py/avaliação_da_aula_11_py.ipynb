```python
import random

# Criar as matrizes 10x10
matriz1 = []
matriz2 = []

for i in range(10):
    linha1 = []
    linha2 = []
    for j in range(10):
        valor1 = random.randint(1, 999)   # números entre 1 e 999
        valor2 = random.randint(1, 999)
        linha1.append(valor1)
        linha2.append(valor2)
    matriz1.append(linha1)
    matriz2.append(linha2)

# Criar a matriz soma (matriz3)
matriz3 = []
for i in range(10):
    linha3 = []
    for j in range(10):
        soma = matriz1[i][j] + matriz2[i][j]
        linha3.append(soma)   # guarda a soma na nova matriz
    matriz3.append(linha3)

# Calcular a média dos valores da matriz3
todos_valores = [valor for linha in matriz3 for valor in linha]
media = sum(todos_valores) / len(todos_valores)
# Mostrar os resultados
print("Matriz 1:")
for linha in matriz1:
    print(linha)

print("\nMatriz 2:")
for linha in matriz2:
    print(linha)

print("\nMatriz Soma (Matriz3):")
for linha in matriz3:
    print(linha)

print("\nMédia dos valores da matriz soma:", media)




```
